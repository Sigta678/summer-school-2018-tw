```python
#Try the other edge finding setting

sobelx = cv2.Sobel(img_RGB,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img_RGB,cv2.CV_64F,0,1,ksize=5)
 
plt.subplot(121),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Set the limit of color we want to find
    
## Red
Range_red1 = np.array([165,100,100])
Range_red2 = np.array([180,255,255]) # <- red will met the high point of H
Range_red3 = np.array([0,100,100])
Range_red4 = np.array([15,255,255])
    
## Yellow
Range_yellow1 = np.array([25,100,100])
Range_yellow2 = np.array([45,255,255])

## Blue  
Range_blue1 = np.array([])
Range_blue2 = np.array([])  
    
## Green 
Range_green1 = np.array([])
Range_green2 = np.array([])  

## White : include all H, lower S, and higher V  
Range_white1 = np.array([]) 
Range_white2 = np.array([])
    
## Black
Range_black1 = np.array([]) 
Range_black2 = np.array([])
```

```python
# Get the mask

red1 = cv2.inRange(img_hsv,Range_red1,Range_red2)
red2 = cv2.inRange(img_hsv,Range_red3,Range_red4)
mask_red = cv2.bitwise_or(red1,red2)

mask_yellow = cv2.inRange(img_hsv,Range_yellow1,Range_yellow2)
    
mask_blue = cv2.inRange()
    
mask_green = cv2.inRange()

mask_white = cv2.inRange()
    
mask_black = cv2.inRange()

# Apply the mask

img_red = cv2.bitwise_and(img_RGB,img_RGB, mask= mask_red)
img_yellow = cv2.bitwise_and(img_RGB,img_RGB, mask= mask_yellow)
img_blue = cv2.bitwise_and()
img_green = cv2.bitwise_and()
img_white = cv2.bitwise_and()
img_black = cv2.bitwise_and()

#Plot your work

plt.subplot(231),plt.imshow(img_red,cmap = 'brg')
plt.title('red'), plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(img_yellow,cmap = 'brg')
plt.title('yellow'), plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(img_blue,cmap = 'brg')
plt.title('blue'), plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(img_green,cmap = 'brg')
plt.title('green'), plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(img_white,cmap = 'brg')
plt.title('white'), plt.xticks([]), plt.yticks([])
plt.subplot(236),plt.imshow(img_black,cmap = 'brg')
plt.title('black'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Some work to make our own color pic  

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3, 3))
ker_white = cv2.dilate(mask_white, kernel)
ker_yellow = cv2.dilate(mask_yellow, kernel)
ker_red = cv2.dilate(mask_red, kernel)
ker_blue = cv2.dilate(mask_blue, kernel)
ker_green = cv2.dilate(mask_green, kernel)
ker_black = cv2.dilate(mask_black, kernel)

x = cv2.cvtColor(ker_red, cv2.COLOR_GRAY2BGR)
x[:,:,0] *= 1
x[:,:,1] *= 0
x[:,:,2] *= 0
x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
plt.subplot(231),plt.imshow(x,cmap = 'brg')
plt.title('red'), plt.xticks([]), plt.yticks([])

x = cv2.cvtColor(ker_yellow, cv2.COLOR_GRAY2BGR)
x[:,:,0] *= 0
x[:,:,1] *= 1
x[:,:,2] *= 1
x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
plt.subplot(232),plt.imshow(x,cmap = 'brg')
plt.title('Yellow'), plt.xticks([]), plt.yticks([])






plt.show()
```
